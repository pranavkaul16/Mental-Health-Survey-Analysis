```python
import pandas as pd
import matplotlib.pyplot as plt

color = {'High': 'red', 'Moderate': 'orange', 'Low': 'green'}
male_data = data[data['Gender'] == 'Male']

stress_counts = male_data['Mental Health Status'].value_counts()

stress_counts = stress_counts.reindex(index=['High', 'Moderate', 'Low'])

plt.figure(figsize=(8, 6)) 
ax = stress_counts.plot(kind='bar', color=[color[level] for level in stress_counts.index])
plt.title('Distribution of Mental Health Statuss among Males')
plt.xlabel('Mental Health Status')
plt.ylabel('Count')
plt.xticks(rotation=0)  
plt.grid(axis='y') 

for i, v in enumerate(stress_counts):
    ax.text(i, v + 0.1, str(v), ha='center', va='bottom')

plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
female_data = data[data['Gender'] == 'Female']

color = {'High': 'red', 'Moderate': 'orange', 'Low': 'green'}
stress_counts = female_data['Mental Health Status'].value_counts()
stress_counts = stress_counts.reindex(index=['High', 'Moderate', 'Low'])

plt.figure(figsize=(8, 6))
ax = stress_counts.plot(kind='bar', color=[color[level] for level in stress_counts.index])
plt.title('Distribution of Stress Levels among females')
plt.xlabel('Stress Level')
plt.ylabel('Count')
plt.xticks(rotation=0) 
plt.grid(axis='y')  

for i, v in enumerate(stress_counts):
    ax.text(i, v + 0.1, str(v), ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

cgpa_bins = [5, 6, 7, 8, 9, 10]  
cgpa_labels = ['5-6', '6-7', '7-8', '8-9', '9-10']  

data['CGPA Range'] = pd.cut(data['CGPA'], bins=cgpa_bins, labels=cgpa_labels)
avg_sleep_hours_by_cgpa = data.groupby('CGPA Range')['Sleep Hours'].mean()

plt.plot(avg_sleep_hours_by_cgpa.index, avg_sleep_hours_by_cgpa.values, marker='o', color='red', label='Average Sleep Hours')

plt.title('Distribution of Sleep Hours and Average Sleep Hours by CGPA Range')
plt.xlabel('Sleep Hours')
plt.ylabel('Frequency / Average Sleep Hours')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('mentalhealth - Sheet1.csv')
color_map = {'High': 'red', 'Moderate': 'orange', 'Low': 'green'}
status_map = {'High': 3, 'Moderate': 2, 'Low': 1}

plt.figure(figsize=(12, 8))

plt.scatter(df['Academic Workload'], df['Mental Health Status'].map(status_map),
            color=df['Academic Workload'].map(color_map), label='Academic Workload', alpha=0.6)

plt.scatter(df['Relationship Problems'], df['Mental Health Status'].map(status_map),
            color=df['Relationship Problems'].map(color_map), label='Relationship Problems', alpha=0.6)

plt.scatter(df['Financial Stress'], df['Mental Health Status'].map(status_map),
            color=df['Financial Stress'].map(color_map), label='Financial Stress', alpha=0.6)

plt.title('Impact of Academic Workload, Relationship Problems, and Financial Stress on Mental Health Status')
plt.xlabel('Level of Stress/Frequency')
plt.ylabel('Mental Health Status (1: Low, 2: Moderate, 3: High)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

age_mental_health_counts = data.groupby('Age')['Mental Health Status'].value_counts().unstack().fillna(0)

age_mental_health_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Mental Health Scores Across Age Groups')
plt.legend(title='Mental Health')
plt.show()
```

```python

academic_year_mental_health_counts = data.groupby(['Academic Year', 'Mental Health Status']).size().unstack(fill_value=0)

mental_health_order = ['Low', 'Moderate', 'High']
academic_year_mental_health_counts = academic_year_mental_health_counts.reindex(columns=mental_health_order)

academic_year_mental_health_counts  = academic_year_mental_health_counts .reindex(index=['Freshman', 'Sophomore', 'Junior','Senior'])

academic_year_mental_health_counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Academic Year')
plt.ylabel('Count')
plt.title('Distribution of Mental Health Status Across Academic Years')
plt.legend(title='Mental Health Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

gender_social_support_counts = data.groupby(['Gender', 'Social Support ']).size().unstack(fill_value=0)

social_support_order = ['Low', 'Moderate', 'High']
gender_social_support_counts = gender_social_support_counts.reindex(columns=social_support_order)

gender_social_support_counts.plot(kind='bar', stacked=True)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Comparison of Gender and Social Support')
plt.legend(title='Social Support')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

variables = ['Academic Workload', 'Financial Stress', 'Relationship Problems']

fig, axes = plt.subplots(nrows=1, ncols=len(variables), figsize=(15, 5))

for i, variable in enumerate(variables):
    gender_variable_counts = data.groupby(['Gender', variable]).size().unstack(fill_value=0)
    
    categories_order = ['Low', 'Moderate', 'High']
    gender_variable_counts = gender_variable_counts.reindex(columns=categories_order)
    
    gender_variable_counts.plot(kind='bar', stacked=True, ax=axes[i])
    axes[i].set_xlabel('Gender')
    axes[i].set_ylabel('Count')
    axes[i].set_title(f'Comparison of Gender and {variable}')
    axes[i].legend(title=variable)
    axes[i].set_xticklabels(['Female', 'Male'], rotation=0)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
coping_strategy_counts = data['Coping Strategies'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(coping_strategy_counts, labels=coping_strategy_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Coping Strategies')
plt.axis('equal')  
plt.show()
```
